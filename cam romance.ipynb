Script for batch updating the callnumbers of a set of item records in the Sierra ILS.

Script uses the [sierra-ils-utils ](https://pypi.org/project/sierra-ils-utils/) Package, created by Ray Voelker to provide a streamlined interface to the Sierra API's.

In this instance the script is taking a csv of item records, containing record numbers, the current call number (merely as a reference) and the associated bib record's 100 field, subfield a, containing the name of the author in last name, first name order.  The library wished to updated all these call numbers to take the form of:

>ROMANCE last_name, first_name

```python
import json
import sierra_ils_utils
import csv
import configparser
```

```python
config = configparser.ConfigParser()
config.read('Y:\\SQL Reports\\creds\\api_info.ini')

'''
.ini file contains url/key/secret for the api in the following form
[api]
base_url = https://[local domain]/iii/sierra-api/v6
client_key = [enter Sierra API key]
client_secret = [enter Sierra API secret]
'''

base_url = config['api']['base_url'] + '/'
#note sierra-ils-utils assumes base_url contains the trailing /, which the file I have been using did not contain so it is appended here
client_key = config['api']['client_key']
client_secret = config['api']['client_secret']

sierra_api = sierra_ils_utils.SierraAPI(base_url,client_key,client_secret)
sierra_api.get('info/token')
```

```python
with open('cam_romance.txt','r') as csv_file:
    csv_reader = csv.DictReader(csv_file, delimiter=',')
    for row in csv_reader:
        print(row['RECORD #(ITEM)'][1:-1] + ' ROMANCE ' + row['100|a'].rstrip(',').rstrip('.'))
        url = base_url + 'items/' + row['RECORD #(ITEM)'][1:-1]
        call_number = 'ROMANCE ' + row['100|a'].rstrip(',').rstrip('.')
        response = sierra_api.session.put(
            url,
            json={
                'callNumbers': [call_number]
            }
        )
        print(response.status_code)
        print(response.text)
```
